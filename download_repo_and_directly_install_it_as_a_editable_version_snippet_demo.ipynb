# 套件安裝

```python
from collections.abc import Collection
# @title 常用參數

# @markdown  要改對應的 project base 路徑, 下面會下載目錄 coqui-ai-TTS-v0.27.2/
PROJ_BASE_PATH = "/content/drive/MyDrive/TEST_TEMP_/taiwan_lang_projXX" # @param {type:"string", placeholder:"project base 路徑"}

# @markdown  check to download coqui-TTS source code for the first time
FirstRun = True # @param {type:"boolean", placeholder:"1st run will download a lot of stuff"}
```

```python
#@title create and change working directory to 專案目錄
!mkdir "$PROJ_BASE_PATH"

%cd {PROJ_BASE_PATH}
%pwd
```

```python
COQUI_TTS_GIT_PATH = "https://github.com/idiap/coqui-ai-TTS.git"
TAG_NAME = "v0.27.2"    # specific version to download
COLON_DIR = f"{PROJ_BASE_PATH}/coqui-ai-TTS-{TAG_NAME}"  #expected destination subdirectory for coqui to save
# print(COLON_DIR)

# !git config --global core.fileMode false   ### ignore File permission bits (executable flag)
# !git config --global core.autocrlf false   ### Then Git DONOT converts LF → CRLF when checking out.
if(FirstRun):
  !rm -rf {COLON_DIR}   ### remove existing download, if any
  !git clone -b "$TAG_NAME" "$COQUI_TTS_GIT_PATH" "$COLON_DIR"
```

```python
#@title change to coqui TTS directory & dump tag to confirm
%cd "$COLON_DIR"
%pwd

!git describe --tags --abbrev=0
```
